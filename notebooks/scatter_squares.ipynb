```python
# Importing Modules
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Graph Rev 1
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(2, 4)
plt.show()
fig.savefig('../outputs/scatter_output1.png')
```

```python
# Graph Rev 2
plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(2, 4, s = 200)

# Setting titles and axes names
ax.set_title('Square Numbers', fontsize = 20)
ax.set_xlabel('Value', fontsize = 14)
ax.set_ylabel('Square of Values', fontsize = 14)

# Set size of the ticks labels
ax.tick_params(axis='both', which='major', labelsize = 14)

plt.show()
fig.savefig('../outputs/scatter_output2.png')
```

```python
# Graph Rev 3
x_values = [1,2,3,4,5]
y_values = [1,4,9,16,25]

plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s = 100)

# Setting titles and axes names
ax.set_title('Square Numbers', fontsize = 20)
ax.set_xlabel('Value', fontsize = 14)
ax.set_ylabel('Square of Values', fontsize = 14)

# Set size of the ticks labels
ax.tick_params(axis='both', which='major', labelsize = 14)

plt.show()
fig.savefig('../outputs/scatter_output3.png')
```

```python
# Graph Rev 4
x_values = range(1, 1001)
y_values = [x**2 for x in x_values]

plt.style.use('seaborn')
fig, ax = plt.subplots()
ax.scatter(x_values, y_values, s = 10)

# Setting titles and axes names
ax.set_title('Square Numbers', fontsize = 20)
ax.set_xlabel('Value', fontsize = 14)
ax.set_ylabel('Square of Values', fontsize = 14)

# Set size of the ticks labels
ax.tick_params(axis='both', which='major', labelsize = 14)

# Set the range for each axis
ax.axis([0, 1100, 0, 1100000])

plt.show()
fig.savefig('../outputs/scatter_output4.png')
```
